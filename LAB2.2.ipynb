```python
import numpy as np
import matplotlib.pyplot as plt

# Input vector
A1 = np.array([[0.05], [0.10]])

# Weight matrices and biases
W2 = np.array([[0.15, 0.20], [0.25, 0.30]])
W3 = np.array([[0.40, 0.45], [0.50, 0.55]])
b2 = np.array([[0.35], [0.35]])
b3 = np.array([[0.60], [0.60]])

# Target output
R = np.array([[0.01], [0.99]])

# Learning rate
alpha = 1

# Activation functions
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def sigmoid_derivative(x):
    return sigmoid(x) * (1 - sigmoid(x))

# Number of epochs
epochs = 10

# Error tracking
errors = np.zeros(epochs)
```

```python
# Training loop
for epoch in range(epochs):
    # Forward pass
    Z2 = np.dot(W2, A1) + b2
    A2 = sigmoid(Z2)

    Z3 = np.dot(W3, A2) + b3
    A3 = sigmoid(Z3)

    # Error deltas
    D3 = (A3 - R) * sigmoid_derivative(Z3)
    D2 = np.dot(W3.T, D3) * sigmoid_derivative(Z2)

    # Gradients
    grad_W3 = np.dot(D3, A2.T)
    grad_b3 = D3
    grad_W2 = np.dot(D2, A1.T)
    grad_b2 = D2

    # Update weights and biases
    W2 = W2 - alpha * grad_W2
    b2 = b2 - alpha * grad_b2
    W3 = W3 - alpha * grad_W3
    b3 = b3 - alpha * grad_b3

    # Calculate total error
    total_error = np.sum((A3 - R) ** 2) / 2
    errors[epoch] = total_error

    # Print epoch info
    print(f"Epoch {epoch + 1}:")
    print("A2:\n", A2)
    print("A3:\n", A3)
    print("Z2:\n", Z2)
    print("Z3:\n", Z3)
    print("D2:\n", D2)
    print("D3:\n", D3)
    print("W2:\n", W2)
    print("W3:\n", W3)
    print("b2:\n", b2)
    print("b3:\n", b3)
    print("-" * 40)
```

```python
# Plot the graph of total errors
plt.figure()
plt.plot(range(1, epochs + 1), errors, '-o')
plt.xlabel('Epochs')
plt.ylabel('Total Error')
plt.title('Total Error vs. Epochs')
plt.grid(True)
plt.show()
```
