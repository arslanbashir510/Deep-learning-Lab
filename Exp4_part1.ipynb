Plot Sigmoids with Different Steepness (w)

```python
import numpy as np
import matplotlib.pyplot as plt
def sigmoid(x, w=1, b=0):
    return 1 / (1 + np.exp(-(w*x + b)))
x = np.linspace(-10, 10, 400)  # Input range from -10 to 10
w_values = [1, 1.5, 3]  # Different steepness (weight) values
b_values = [0, 1, 2]    # Different bias (position) values
colors = ['b', 'orange', 'g']  # Plot colors
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
for i, w in enumerate(w_values):
    plt.plot(x, sigmoid(x, w=w, b=0), color=colors[i], label=f'w = {w}, b = 0', linestyle='--')
plt.title('Effect of w (Steepness) on Sigmoid')
plt.xlabel('x')
plt.ylabel('Ïƒ(x)')
plt.legend()
plt.grid()
```

Display Plots

```python
import numpy as np
import matplotlib.pyplot as plt
def sigmoid(x, w=1, b=0):
    return 1 / (1 + np.exp(-(w*x + b)))
x = np.linspace(-10, 10, 400)  # Input range from -10 to 10
w_values = [1, 1.5, 3]  # Different steepness (weight) values
b_values = [0, 1, 2]    # Different bias (position) values
colors = ['b', 'orange', 'g']  # Plot colors
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
for i, w in enumerate(w_values):
    plt.plot(x, sigmoid(x, w=w, b=0), color=colors[i], label=f'w = {w}, b = 0', linestyle='--')
plt.title('Effect of w (Steepness) on Sigmoid')
plt.xlabel('x')
plt.ylabel('Ïƒ(x)')
plt.legend()
plt.grid()
plt.subplot(1, 2, 2)
for i, b in enumerate(b_values):
    plt.plot(x, sigmoid(x, w=1, b=b), color=colors[i], label=f'w = 1, b = {b}', linestyle='dotted')
plt.title('Effect of b (Position) on Sigmoid')
plt.xlabel('x')
plt.ylabel('Ïƒ(x)')
plt.legend()
plt.grid()
plt.tight_layout()
plt.show()
```
